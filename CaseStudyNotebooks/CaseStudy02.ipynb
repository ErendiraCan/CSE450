```python
# New Column: Predictions
market_data['predictions'] = np.where((market_data.poutcome == 1), 1,
             np.where((market_data.poutcome == 2), 0,
             0))
```

The Model: Decision Tree

```python
# scikit learn froms
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import GradientBoostingClassifier
from sklearn import metrics
```

```python
# Training Data and Testing Data
X_pred = market_data.drop(market_data.filter(regex = 'housing|pdays|emp.var.rate|cons.price.idx|cons.conf.idx|euribor3m|nr.employed|predictions|poutcome')
.columns, axis = 1)
y_pred = market_data.filter(regex = "predictions")
```

```python
#| See Data
X_train, X_test, y_train, y_test = train_test_split(X_pred, y_pred, test_size = .34, random_state = 76) 
y_test
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
y_probs = clf.predict_proba(X_test)
```

```python
print(metrics.classification_report(y_pred, y_test))
```

```python
df_features = pd.DataFrame(
    {'f_names': X_train.columns, 
    'f_values': clf.feature_importances_}).sort_values('f_values', ascending = False)

Metrics = (alt.Chart(df_features.query('f_values > .011'))
    .encode(
        alt.X('f_values'),
        alt.Y('f_names', sort = '-x'))
    .mark_bar())

Metrics
```
