```python
# Part 1: Enter the code below to import Pandas according to the 
# conventional method.
import pandas as pd
```

```python
# Part 2: Load the dataset into a Pandas dataframe.
market_data = pd.read_csv("https://raw.githubusercontent.com/byui-cse/cse450-course/master/data/bank_holdout_test.csv")
```

```python
# Then, explore the data by seeing what the first few rows look like. 
market_data.head()
```

```python
# Remove all NaN values
market_data = market_data.dropna()
```

```python
# See the values inside job column
job_counts = market_data['job'].value_counts()

# Print the counts
print(job_counts)
```

```python
# Convert str to numerical values: JOB

job_map = {
    'admin.': 1,
    'blue-collar': 2,
    'entrepreneur': 3,
    'housemaid': 4,
    'management': 5,
    'retired': 6,
    'self-employed': 7,
    'services': 8,
    'student': 9,
    'technician': 10,
    'unemployed': 11,
    'unknown': 12
}

market_data['job'] = market_data['job'].map(job_map)

market_data.head()
```

```python
# See the values inside marital column
marital_counts = market_data['marital'].value_counts()

# Print the counts
print(marital_counts)
```

```python
# Convert str to numerical values: MARITAL

marital_map = {
    'married': 1,
    'single': 2,
    'divorced': 3,
    'unknown': 4
}

market_data['marital'] = market_data['marital'].map(marital_map)

market_data.head()
```

```python
# See the values inside education column
education_counts = market_data['education'].value_counts()

# Print the counts
print(education_counts)
```

```python
# Convert str to numerical values: EDUCATION

education_map = {
    'university.degree': 1,
    'high.school': 2,
    'basic.9y': 3,
    'professional.course': 4,
    'basic.4y': 5,
    'basic.6y': 6,
    'unknown': 7,
    'illiterate': 8
}

market_data['education'] = market_data['education'].map(education_map)

market_data.head()
```

```python
# See the values inside contact column
contact_counts = market_data['contact'].value_counts()

# Print the counts
print(contact_counts)
```

```python
# Convert str to numerical values: CONTACT

contact_map = {
    'cellular': 1,
    'telephone': 2
}

market_data['contact'] = market_data['contact'].map(contact_map)

market_data.head()
```

```python
# See the values inside loan column
loan_counts = market_data['loan'].value_counts()

# Print the counts
print(loan_counts)
```

```python
# Convert str to numerical values: LOAN

loan_map = {
    'yes': 1,
    'no': 2,
    'unknown': 3
}

market_data['loan'] = market_data['loan'].map(loan_map)

market_data.head()
```
